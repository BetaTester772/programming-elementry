```python
import matplotlib.pyplot as plt
import math

x_raw = list(map(int, production.loc[total_production, '2019']))
y_raw = list(map(int, enterprise.iloc[1:, 1]))

print(x_raw)
print(y_raw)

x = x_raw
y = y_raw


class Neuron:
    def __init__(self):
        self.w = 1.0
        self.b = 1.0

    def forpass(self, x):
        y_hat = x * self.w + self.b
        return y_hat

    def backprop(self, x, err):
        w_grad = x * err
        b_grad = 1 * err
        return w_grad, b_grad

    def fit(self, x, y, epochs=100):
        for i in range(epochs):
            for x_i, y_i in zip(x, y):
                y_hat = self.forpass(x_i)
                err = -(y_i - y_hat)
                w_grad, b_grand = self.backprop(x_i, err)
                if math.isinf(self.w - w_grad) or math.isinf(self.b - b_grand):
                    print(False, self.w, self.b, w_grad, b_grand, i)
                    return
                self.w -= w_grad
                self.b -= b_grand
                print(self.w, self.b, w_grad, b_grand, i)


nueron = Neuron()
nueron.fit(x, y)

print(nueron.w, nueron.b)

print(True, x, y)

plt.scatter(x, y)
pt1 = (-0.1, -0.1 * nueron.w + nueron.b)
pt2 = (0.15, 0.15 * nueron.w + nueron.b)
print(pt1, pt2)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.ticklabel_format(style='plain')
plt.show()
```
