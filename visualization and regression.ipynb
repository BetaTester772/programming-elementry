```python
# 전체

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

enterprise = pd.read_csv('로봇산업_기업_현황_20221130123620.csv')
sales = pd.read_csv('로봇산업_매출_현황_20221130123612.csv')
production = pd.read_csv('로봇산업_생산_현황_20221130123628.csv')

total_production = [0, 1, 10, 19, 25, 32, 36, 42]

display(enterprise)
display(sales)
display(production.iloc[total_production])

plt.figure(figsize=(21, 7))

plt.subplot(1, 2, 1)

# 2019
x = enterprise.loc[1:, '2019']
y = sales.loc[1:, '2019']

x = np.array(x)
y = np.array(y)
plt.scatter(x, y)

for i in range(0, 7):
    plt.text(x[i], y[i], sales['업종별(1)'][i + 1], fontsize=9, horizontalalignment='left', verticalalignment='bottom',
             rotation=90)

x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

model = LinearRegression()
model.fit(x, y)
print(model.coef_, model.intercept_)
print(model.score(x, y))
plt.plot(x, model.predict(x))

# 2020
x = []
y = []

x = enterprise.loc[1:, '2020']
y = sales.loc[1:, '2020']

x = np.array(x)
y = np.array(y)

for i in range(0, 7):
    plt.text(x[i], y[i], sales['업종별(1)'][i + 1], fontsize=9, horizontalalignment='left', verticalalignment='bottom',
             rotation=90)

x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

model = LinearRegression()
model.fit(x, y)
print(model.coef_, model.intercept_)
print(model.score(x, y))

plt.xlabel('생산')
plt.ylabel('매출')
plt.title("2019-2020 로봇산업 매출-기업 현황")
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'orange')
plt.legend(['2019', '2020', '2019', '2020'])

#
plt.subplot(1, 2, 2)

# 2019
x = []
for i in range(7):
    x.append(int(production.iloc[total_production[i], 2]))
y = sales['2019'][1:]

x = np.array(x)
y = np.array(y)

for i in range(7):
    plt.text(x[i], y[i], sales['업종별(1)'][i + 1], fontsize=9, horizontalalignment='left', verticalalignment='bottom',
             rotation=90)

x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
print(model.coef_, model.intercept_)
print(model.score(x, y))

plt.scatter(x, y)
plt.plot(x, model.predict(x))

# 2020
x = []
for i in range(7):
    x.append(int(production.iloc[total_production[i], 3]))
y = sales['2020'][1:]

x = np.array(x)
y = np.array(y)

for i in range(7):
    plt.text(x[i], y[i], sales['업종별(1)'][i + 1], fontsize=9, horizontalalignment='left', verticalalignment='bottom',
             rotation=90)

x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
print(model.coef_, model.intercept_)
print(model.score(x, y))

plt.title("2019-2020 로봇산업 매출-생산 현황")
plt.scatter(x, y)
plt.xlabel('생산')
plt.ylabel('매출')
plt.plot(x, model.predict(x))
plt.legend(['2019', '2020', '2019', '2020'])
plt.show()
```
